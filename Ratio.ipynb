```python
import re
import pandas as pd
```

```python
parsed_nomenclature = pd.read_excel("parsed_nomenclature.xlsx").dropna(how="all").fillna("")
parsed_nomenclature.head()
```

```python
test_nomenclature = pd.read_excel("test_nomenclature.xlsx").dropna(how="all").fillna("").drop_duplicates()
test_nomenclature.head()
```

```python
characteristics = {
    "Производитель/ Бренд": "producer",
    "Модель/ Партномер/ Название/ ГОСТ(Стандарт)": "name",
    "Назначение/ Принадлженость": "destination",
    "Классификатор": "model",
    "Тип/ Исполнение/ Конструкция/ Разновидность": "type",
    "Материал\n(Выбор из списка)": "material",
    "Цвет (Выбор из списка)": "color",
    "Размеры": "size",
    "Вес/Объем": "weight",
    "Физическая величина": "physics",
    "Доп признак:\n- Количество/Количество в упаковке": "packaging"
}
characteristics
```

```python
total_count = 0
parsed_count = 0

for index, row in parsed_nomenclature.iterrows():
    test_rows = test_nomenclature.loc[test_nomenclature["INPUT:text"] == row["Наименование ТМЦ"]]
    if not test_rows.empty:
        for column in characteristics.keys():
            if row[column] != "" and row[column] is not None:
                test_cells = test_rows.loc[test_nomenclature["OUTPUT:characteristic"] == characteristics[column]]["INPUT:part"]
                if not test_cells.empty:
                    parsed_values = re.split(r' |\n|;|\\(|\\)', row[column])
                    for parsed_value in parsed_values:
                        print(parsed_value)
                    print("---")
                    for test_cell in test_cells:
                        print(test_cell)
                    total_count += 1
                    for parsed_value in parsed_values:                        
                        if any(test_cell == parsed_value for test_cell in test_cells if test_cell is not None and test_cell != ""):
                            parsed_count += 1
                            break
                else:
                    print("empty")
                print("---------------")
print(parsed_count / total_count)
```
