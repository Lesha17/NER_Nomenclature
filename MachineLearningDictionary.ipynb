```python
import re
import pandas as pd
import numpy as np
```

```python
nomenclature = pd.read_excel("nomenclature.xlsx")
print(nomenclature.shape)
nomenclature = nomenclature.append(pd.read_excel("nomenclature2.xlsx"))
print(nomenclature.shape)
nomenclature.head()
```

```python
nomenclature_patterns = pd.read_excel("nomenclature_patterns.xlsx")
nomenclature_patterns.head()
```

```python
nomenclature_characteristics = nomenclature.join(nomenclature_patterns.set_index("Наименование целевого ОЗМ"),
                                                on="Наименование целевого ОЗМ")
nomenclature_characteristics.head()
```

```python
columns = nomenclature_characteristics.columns
columns
```

```python
def split_string(string):
    return re.split(' |\n|;|,', str(string))
```

```python
dictionary = set()
for index, item in nomenclature_characteristics.iterrows():
    for column in columns:
        dictionary = dictionary.union(set(split_string(item[column])))

dictionary
```

```python
with open('output2.txt', 'w') as f:
    for dictinary_item in dictionary:
        f.write("%s\n" % dictinary_item)
```
